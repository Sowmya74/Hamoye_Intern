```python
# Distribution of Food Supply over the years
plt.figure(figsize=(14, 10))
sns.lineplot(data=food_df, x='Area', y='Y2014', label='2014')
sns.lineplot(data=food_df, x='Area', y='Y2015', label='2015')
sns.lineplot(data=food_df, x='Area', y='Y2016', label='2016')
sns.lineplot(data=food_df, x='Area', y='Y2017', label='2017')
sns.lineplot(data=food_df, x='Area', y='Y2018', label='2018')
plt.title('Food Supply Over the Years')
plt.xlabel('Area')
plt.ylabel('Food Supply (kcal/capita/day)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python

# Bar plot comparing total population across different areas for each year
plt.figure(figsize=(12, 6))
sns.barplot(data=food_df, x='Area', y='Y2014', color='skyblue', alpha=0.7, label='2014')
sns.barplot(data=food_df, x='Area', y='Y2015', color='salmon', alpha=0.7, label='2015')
sns.barplot(data=food_df, x='Area', y='Y2016', color='green', alpha=0.7, label='2016')
sns.barplot(data=food_df, x='Area', y='Y2017', color='orange', alpha=0.7, label='2017')
sns.barplot(data=food_df, x='Area', y='Y2018', color='purple', alpha=0.7, label='2018')
plt.title('Total Population Across Different Areas (2014-2018)')
plt.xlabel('Area')
plt.ylabel('Total Population (in 1000 persons)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Box plot to visualize the distribution of food supply across different years
plt.figure(figsize=(10, 6))
sns.boxplot(data=food_df[['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']], orient='h')
plt.title('Distribution of Food Supply Across Different Years')
plt.xlabel('Food Supply (kcal/capita/day)')
plt.ylabel('Year')
plt.tight_layout()
plt.show()

# Violin plot to show the distribution of protein supply across different years
plt.figure(figsize=(10, 6))
sns.violinplot(data=food_df[['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']], orient='h')
plt.title('Distribution of Protein Supply Across Different Years')
plt.xlabel('Protein Supply (g/capita/day)')
plt.ylabel('Year')
plt.tight_layout()
plt.show()

# Exclude non-numeric columns
numeric_columns = food_df.select_dtypes(include=['float64', 'int64']).columns

# Heatmap to visualize the correlation between numeric variables
plt.figure(figsize=(10, 6))
sns.heatmap(food_df[numeric_columns].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

### **Question 1: Which year had the least correlation with ‘Element Code’?**

```python
col = 'Element Code'
years = ['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']

# Calculating the correlation between the specific column and the years
correlation_with_years = food_df[years].corrwith(food_df[col])

# Bar Plot the correlation
plt.figure(figsize=(8, 5))
#sns.barplot(x=correlation_with_years.index, y=correlation_with_years.values, palette='coolwarm')
#Got a Future Warning about Hue
sns.barplot(x=correlation_with_years.index, y=correlation_with_years.values, hue=correlation_with_years.index, palette='magma', legend=False)
plt.title(f'Correlation of {col} with Years (2014-2018)')
plt.xlabel('Year')
plt.ylabel('Correlation')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### **Question 3 Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the highest sum in 2017?**

```python
data = food_df.groupby('Area')['Y2017'].sum()

#area with the highest sum
highest_total_area = data.idxmax()

highest_total = data.max()

print("Area with the highest sum in Y2017:", highest_total_area)
print("Highest sum value in Y2017:", highest_total)
```
